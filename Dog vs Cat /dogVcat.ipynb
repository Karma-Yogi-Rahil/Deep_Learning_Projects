```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import os
import random
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
training_dir = '/content/train_224'
validate_dir = '/content/valid_224'
test_dir = '/content/test_224'
```

```python
training_dog_dir = os.path.join(training_dir,'dog')
training_cat_dir = os.path.join(training_dir, 'cat')
print(len(os.listdir(training_dog_dir)))
print(len(os.listdir(training_cat_dir)))
```

```python
def view_random_image(target_dir, target_class ,row , col,size):
  fig = plt.gcf()
  fig.set_size_inches(col * size, row * size)
  target_folder = os.path.join(training_dir,target_class)
  random_image = random.sample(os.listdir(target_folder), col)
  print(f"\n {random_image} \n")
  for i,images in enumerate(random_image):
    
    img = mpimg.imread(target_folder + '/' + random_image[i])
    #print(target_folder + '/' + random_image[i])
    plt.subplot(1, col, i+1)
    
    plt.imshow(img)
    plt.title(target_class)
    plt.axis('off')
  plt.show()
```

```python
fig_size = 3
```

```python
print(f"\n--------------------------------------------Training-----------------------------------------------")
view_random_image(training_dir,'dog' , row =1 ,col =5,size = fig_size)
view_random_image(training_dir,'cat' , row =1 ,col =5,size = fig_size)
print(f"\n\n--------------------------------------------Validate------------------------------------------------------")
view_random_image(validate_dir,'dog' , row =1 ,col =5,size = fig_size)
view_random_image(validate_dir,'cat' , row =1 ,col =5,size = fig_size)
```
